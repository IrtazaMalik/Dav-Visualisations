```python
import pandas as pd

# Load the datasets
data1 = pd.read_csv('incidence-of-tuberculosis-sdgs.csv')
data2 = pd.read_csv('tuberculosis-deaths-by-age.csv')
```

```python
# Check for missing values
data1.isnull().sum()
data2.isnull().sum()

# Fill missing values (if appropriate, here filling with the mean for example)
data1['Estimated incidence of all forms of tuberculosis'] = data1['Estimated incidence of all forms of tuberculosis'].fillna(data1['Estimated incidence of all forms of tuberculosis'].mean())
data2.fillna(0, inplace=True)  # Assuming missing death values can be treated as zero
```

```python
data1['Year'] = data1['Year'].astype(int)
data2['Year'] = data2['Year'].astype(int)
```

```python
data1['Estimated incidence of all forms of tuberculosis'] = data1['Estimated incidence of all forms of tuberculosis'].fillna(data1['Estimated incidence of all forms of tuberculosis'].mean())
```

```python
death_columns = [
    'Deaths - Tuberculosis - Sex: Both - Age: 70+ years (Number)',
    'Deaths - Tuberculosis - Sex: Both - Age: 50-69 years (Number)',
    'Deaths - Tuberculosis - Sex: Both - Age: 15-49 years (Number)',
    'Deaths - Tuberculosis - Sex: Both - Age: 5-14 years (Number)',
    'Deaths - Tuberculosis - Sex: Both - Age: Under 5 (Number)'
]
data2[death_columns] = data2[death_columns].fillna(0)
```

```python
data2['Year'] = data2['Year'].astype(int)
```

```python
# Merge data1 and data2 on 'Entity', 'Code', and 'Year'
merged_data = pd.merge(data1, data2, on=['Entity', 'Code', 'Year'], how='inner')
```

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('cleaned_tb_data.csv')

# Check the first few rows to understand the structure
print(df.head())

# For simplicity, let's focus on columns that are important for the graph
# Columns: 'Entity', 'Estimated incidence of all forms of tuberculosis', and death counts for different age groups
# You can add more columns for detailed relationships (like age group death counts).

# Step 1: Clean and process the data
# Select relevant columns
df_cleaned = df[['Entity', 'Code', 'Year', 'Estimated incidence of all forms of tuberculosis', 
                 'Deaths - Tuberculosis - Sex: Both - Age: 70+ years (Number)', 
                 'Deaths - Tuberculosis - Sex: Both - Age: 50-69 years (Number)', 
                 'Deaths - Tuberculosis - Sex: Both - Age: 15-49 years (Number)', 
                 'Deaths - Tuberculosis - Sex: Both - Age: 5-14 years (Number)', 
                 'Deaths - Tuberculosis - Sex: Both - Age: Under 5 (Number)']]

# Step 2: Create a relationship (edge) based on the similarity of incidence rates
# For simplicity, let's use a threshold for similarity (e.g., countries with incidence rates within ±10% of each other)

# Calculate the pairwise difference of incidence rates between countries
def create_relationships(df):
    edges = []
    nodes = df['Entity'].unique()
    
    for i, country1 in enumerate(nodes):
        for j, country2 in enumerate(nodes):
            if i < j:
                # Get incidence rates for both countries
                incidence1 = df[df['Entity'] == country1]['Estimated incidence of all forms of tuberculosis'].mean()
                incidence2 = df[df['Entity'] == country2]['Estimated incidence of all forms of tuberculosis'].mean()
                
                # Check if the incidence rates are similar (within 10% of each other)
                if abs(incidence1 - incidence2) / min(incidence1, incidence2) < 0.1:
                    edges.append({'source': country1, 'target': country2, 'value': 1})
    
    return edges

# Generate the relationships (edges)
edges = create_relationships(df_cleaned)

# Generate the nodes list from the unique countries
nodes = [{'id': country} for country in df_cleaned['Entity'].unique()]

# Save the data to a JSON format (for D3.js use)
import json

graph_data = {
    'nodes': nodes,
    'links': edges
}

# Save to a JSON file
with open('graph_data.json', 'w') as outfile:
    json.dump(graph_data, outfile)
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
import time

# Load the dataset
data = pd.read_csv("cleaned_tb_data.csv")

# Extract unique country names
countries = data['Entity'].unique()
print(f"Unique countries found: {len(countries)}")

# Initialize geolocator
geolocator = Nominatim(user_agent="geoapi")

# Function to get latitude and longitude
def get_lat_lon(country):
    try:
        location = geolocator.geocode(country)
        if location:
            return pd.Series([location.latitude, location.longitude])
        else:
            return pd.Series([None, None])
    except Exception as e:
        print(f"Error fetching data for {country}: {e}")
        return pd.Series([None, None])

# Apply the function to each country
lat_lon_data = pd.DataFrame(countries, columns=['Entity'])
lat_lon_data[['Latitude', 'Longitude']] = lat_lon_data['Entity'].apply(get_lat_lon)

# Merge lat/lon data with the original dataset
merged_data = pd.merge(data, lat_lon_data, on='Entity', how='left')

# Save the augmented dataset
merged_data.to_csv("cleaned_tb_data_with_lat_lon.csv", index=False)
print("Saved the augmented dataset as cleaned_tb_data_with_lat_lon.csv")
```

```python
import pandas as pd
import json

# Load the dataset
df = pd.read_csv("cleaned_tb_data_with_lat_lon.csv")

# Inspect the first few rows to understand the structure of the data
df.head()

# Create a hierarchical format
def create_hierarchy(df):
    # Create a dictionary to store the hierarchical data
    hierarchy = {"name": "Root", "children": []}

    # Iterate through the dataset and organize the data into hierarchy
    for _, row in df.iterrows():
        # Find or create the country node
        country_node = next((item for item in hierarchy["children"] if item["name"] == row["Entity"]), None)
        if not country_node:
            country_node = {"name": row["Entity"], "children": []}
            hierarchy["children"].append(country_node)

        # Find or create the year node for the country
        year_node = next((item for item in country_node["children"] if item["name"] == str(row["Year"])), None)
        if not year_node:
            year_node = {"name": str(row["Year"]), "children": []}
            country_node["children"].append(year_node)

        # Add the tuberculosis data as a leaf node (e.g., Incidence)
        year_node["children"].append({
            "name": f"Incidence: {row['Estimated incidence of all forms of tuberculosis']}",
            "value": row["Estimated incidence of all forms of tuberculosis"],
            "relationshipType": row["Entity"]  # You can change this to another column like 'Code'
        })
    
    return hierarchy

# Convert the dataframe to hierarchical data
hierarchical_data = create_hierarchy(df)

# Output the hierarchical data in JSON format for Sunburst
hierarchical_json = json.dumps(hierarchical_data, indent=2)

# Save the hierarchical data to a JSON file
with open("hierarchical_tb_data.json", "w") as f:
    f.write(hierarchical_json)

# Display the hierarchical data
hierarchical_data
```

```python
import pandas as pd
import json

# Load the CSV data into a pandas DataFrame
df = pd.read_csv('cleaned_tb_data_with_lat_lon.csv')

# Initialize an empty dictionary to hold the hierarchical structure
data = {}

# Iterate through the DataFrame and organize the data into a nested structure
for _, row in df.iterrows():
    country = row['Entity']
    year = row['Year']
    
    if country not in data:
        data[country] = {}
    
    # If year doesn't exist, initialize it
    if year not in data[country]:
        data[country][year] = {
            'incidence': row['Estimated incidence of all forms of tuberculosis'],
            'deaths': {
                '70+': row['Deaths - Tuberculosis - Sex: Both - Age: 70+ years (Number)'],
                '50-69': row['Deaths - Tuberculosis - Sex: Both - Age: 50-69 years (Number)'],
                '15-49': row['Deaths - Tuberculosis - Sex: Both - Age: 15-49 years (Number)'],
                '5-14': row['Deaths - Tuberculosis - Sex: Both - Age: 5-14 years (Number)'],
                'Under 5': row['Deaths - Tuberculosis - Sex: Both - Age: Under 5 (Number)']
            }
        }

# Convert the data dictionary into JSON format
json_data = json.dumps(data, indent=4)

# Save the JSON data to a file
with open('tb_data_hierarchical.json', 'w') as f:
    f.write(json_data)

print("Data has been converted and saved to 'tb_data_hierarchical.json'")
```
